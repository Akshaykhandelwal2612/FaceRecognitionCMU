```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
directories=os.listdir()
directories
```

```python
from PIL import Image
```

```python
count=0
for i in directories:
  os.chdir(i)
  li=os.listdir()
  for j in li:
    count+=1
  os.chdir("../")
print(count)
    
```

```python
x_train=np.zeros((64*64,count))
y_train=np.zeros((1,count))
```

```python
os.chdir("..\\")
```

```python
os.getcwd()
```

```python
index=0
for i in directories:
  os.chdir(i)
  li=os.listdir()
  for j in li:
    img=Image.open(j)
    img=img.convert("L")
    img=img.resize((64,64))
    s=np.asarray(img)
    s=s.flatten()
    x_train[:,index]=s
    y_train[0,index]=float(i[1:])
    index+=1
  os.chdir("../")
```

```python
x_train=x_train/255.0
```

```python
os.chdir("..//")
os.chdir("test")
os.getcwd()
```

```python
x_test=np.zeros((64*64,35000))
```

```python
images=os.listdir()
index=0
for j in images:
    img=Image.open(j)
    img=img.convert("L")
    img=img.resize((64,64))
    s=np.asarray(img)
    s=s.flatten()
    x_test[:,index]=s
    index+=1
```

```python
x_test=x_test/255.0
```

```python
from tensorflow import keras
```

```python
x_train=x_train.T
x_train.shape
```

```python
x_test=x_test.T
x_test.shape
```

```python
y_train=y_train.flatten().astype('int32')
y_train.shape
```

```python
y_train
```

```python
x_train = x_train.reshape(-1,64,64,1)
x_test = x_test.reshape(-1,64,64,1)
x_train.shape,x_test.shape
```

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(32,3, activation='relu', input_shape=(64,64,1)),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling2D(pool_size=(3, 3)),

    keras.layers.Flatten(),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(9279,activation='softmax')  # activation change
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', # Loss
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train, y_train,epochs=25)
```

```python
results = model.predict(x_test)
results
```

```python
results.shape
```

```python
results = np.argmax(results,axis = 1)
```

```python
results=results+1
```

```python
li=list()
for i in results:
    s="n"
    i=str(i)
    x=len(i)
    s=s+(6-x)*"0"+i
    li.append(s)
li
```

```python
df["id"]=images
```
